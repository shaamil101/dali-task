```python
# Importing Pandas module
import pandas as pd
# Reading JSON file
df = pd.read_json("New Document 1.json")
df
```

```python
# Dropping rows with NaN values
df = df.dropna()
df
```

```python
# List of all column names
list(df.columns)
```

```python
# Dropping unwanted columns
df = df.drop(['year','gender','heightInches','happiness','caffeineRating','affiliated','numOfLanguages','gymPerWeek','phoneType'], axis=1)
```

```python
# Printing Data Set
df
```

```python
# Plotting a scatterplot using Bokeh 
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row 
from bokeh.plotting import figure


p = figure(title = "Socials vs Alcohol Drinks")
p.circle('alcoholDrinksPerWeek','socialDinnerPerWeek',source=df,fill_alpha=0.2, size=10)

show(p)
```

```python
# Correlation Matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# Importing Seaborn Module 
import seaborn as sns
ax = sns.heatmap(corr_matrix, vmin=0, vmax=1)
```
